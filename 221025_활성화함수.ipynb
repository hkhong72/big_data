```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid_function(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-5, 5)
y = sigmoid_function(x)
plt.plot(x, y)
plt.show()
```

```python
def tanh_function(x):
    return np.tanh(x)
```

```python
x = np.linspace(-5, 5)
y = tanh_function(x)
plt.plot(x, y)
plt.show()
```

```python
def relu_function(x):
    return np.where(x <= 0, 0, x)

x = np.linspace(-5, 5)
y = relu_function(x)
plt.plot(x, y)
plt.show()
```

```python
def leaky_relu_function(x):
    return np.where(x <= 0, 0.01*x, x)

x = np.linspace(-5, 5)
y = leaky_relu_function(x)
plt.plot(x, y)
plt.show()
```
