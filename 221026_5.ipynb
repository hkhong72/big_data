```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
# KMP에러 나는걸방지
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
```

```python
df = pd.read_csv('data/sonar3.csv', header=None)
```

```python
df.head() #일반돌멩이 경우 : 0, 광석일 경우 : 1
```

```python
df[60].value_counts()
```

```python
X = df.iloc[:, :60]
y = df.iloc[:, 60]
```

```python
from tensorflow.keras import models, layers
```

```python
model = models.Sequential()
model.add(layers.Dense(24, input_dim=60, activation='relu'))
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.summary()
```

```python
# w + b ?
# (60 x w) * (w x 24) + b(24) == 1464
# (24 x w) * (w x 10) + b(10) == 250
# (10 x w) * (w x 1) + b(1) = 11 
# -> 총191 
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 모델 실행
history = model.fit(X, y, epochs=200, batch_size=10) # accuracy가 1.000 이 나와도 완벽하지않음 -> 과적합 -> 학습셋과 테스트셋을 구분해야 함
```
